```python
df=df.sample(frac=1).reset_index(drop=True)
df.head()
```

```python
y=df['Result']
X=df.drop('Result',axis=1)
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)
X_train.shape, X_test.shape
```

```python
model=[]
train=[]
test=[]
def results(m,a,b):
    model.append(m)
    train.append(round(a,3))
    test.append(round(b,3))
```

```python
##### Random Forest Model - m1 #####
from sklearn.ensemble import RandomForestClassifier
m1=RandomForestClassifier(max_depth=5)
m1.fit(X_train,y_train)
y_test_m1=m1.predict(X_test)
y_train_m1=m1.predict(X_train)
from sklearn.metrics import accuracy_score
train_m1=accuracy_score(y_train,y_train_m1)
test_m1=accuracy_score(y_test, y_test_m1)
print("Random Forest training data Accuracy: {:.2f}".format(train_m1))
print("Random Forest test data Accuracy: {:.2f}".format(test_m1))
```

```python
##### SVM - Support Vector Machine Model - m2 ######
from sklearn.svm import SVC
m2=SVC(kernel='linear',C=1.0, random_state=12)
m2.fit(X_train,y_train)
y_test_m2=m2.predict(X_test)
y_train_m2=m2.predict(X_train)
train_m2=accuracy_score(y_train,y_train_m2)
test_m2=accuracy_score(y_test, y_test_m2)
print("SVM training data Accuracy: {:.2f}".format(train_m2))
print("SVM test data Accuracy: {:.2f}".format(test_m2))
```

```python
##### Decision Tree model - m3 #####
from sklearn.tree import DecisionTreeClassifier
m3 = DecisionTreeClassifier(max_depth = 5)
m3.fit(X_train, y_train)
y_test_m3=m3.predict(X_test)
y_train_m3=m3.predict(X_train)
train_m3=accuracy_score(y_train,y_train_m3)
test_m3=accuracy_score(y_test, y_test_m3)
print("Decision tree training data Accuracy: {:.2f}".format(train_m3))
print("Decision tree test data Accuracy: {:.2f}".format(test_m3))
```

```python
###### XGBoost Model - m4 ######
from xgboost import XGBClassifier
m4 = XGBClassifier(learning_rate=0.4,max_depth=7)
m4.fit(X_train, y_train)
y_test_m4=m4.predict(X_test)
y_train_m4=m4.predict(X_train)
train_m4=accuracy_score(y_train,y_train_m4)
test_m4=accuracy_score(y_test, y_test_m4)
print("XGBOOST training data Accuracy: {:.2f}".format(train_m4))
print("XGBOOST test data Accuracy: {:.2f}".format(test_m4))
```

```python
results('Random Forest', train_m1, test_m1)
results('SVM', train_m2, test_m2)
results('Decision Tree', train_m3, test_m3)
results('XGBoost', train_m4, test_m4)
```
